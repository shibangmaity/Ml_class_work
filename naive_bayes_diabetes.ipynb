```python
import pandas as pd
import numpy as np
import math
```

```python
# Load dataset (keep diabetes.csv in same folder)
df = pd.read_csv("diabetes.csv")
df.head()
```

```python
# Separate features and target
X = df.drop("Outcome", axis=1)
y = df["Outcome"]
```

```python
# Split data by class
df_yes = df[df["Outcome"] == 1]
df_no  = df[df["Outcome"] == 0]

P_yes = len(df_yes) / len(df)
P_no  = len(df_no) / len(df)

print(P_yes, P_no)
```

```python
# Mean and Standard Deviation
mean_yes = df_yes.drop("Outcome", axis=1).mean()
std_yes  = df_yes.drop("Outcome", axis=1).std()

mean_no  = df_no.drop("Outcome", axis=1).mean()
std_no   = df_no.drop("Outcome", axis=1).std()
```

```python
# Gaussian Probability Function
def gaussian(x, mean, std):
    exponent = math.exp(-((x - mean) ** 2) / (2 * std ** 2))
    return (1 / (math.sqrt(2 * math.pi) * std)) * exponent
```

```python
# Take one test sample (first row)
test = X.iloc[0]

# Calculate posterior probabilities
P_yes_final = P_yes
P_no_final  = P_no

for col in X.columns:
    P_yes_final *= gaussian(test[col], mean_yes[col], std_yes[col])
    P_no_final  *= gaussian(test[col], mean_no[col], std_no[col])
```

```python
print("P(Yes | X) =", P_yes_final)
print("P(No  | X) =", P_no_final)
```

```python
# Final Prediction
if P_yes_final > P_no_final:
    print("Prediction: DIABETIC")
else:
    print("Prediction: NOT DIABETIC")
```
