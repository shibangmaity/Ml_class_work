```python
x1 = [2104, 1416, 1532, 852, 1800, 2000, 1200, 950, 1600, 1100]
x2 = [5, 3, 3, 2, 4, 4, 2, 2, 3, 2]
x3 = [1, 2, 2, 1, 2, 1, 1, 1, 2, 1]
x4 = [45, 40, 30, 36, 42, 50, 35, 30, 48, 28]
y = [460, 232, 315, 178, 380, 420, 200, 190, 350, 210]
#Values of x1 , x2 , x3 , x4 , y
#Target is y price #

X = np.array([x1 , x2 , x3 , x4])
X = X.T
y = np.array(y)
y = y.T
X = pd.DataFrame(X)
y = pd.DataFrame(y)

df = pd.concat([X , y] , axis = 1)
print(df.head())

X = df.iloc[: , :-1]
y = df.iloc[: , -1]
     
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler # Add this import

# Extract features and target from df, ensuring X does not have a bias column at this stage
X_features = df.iloc[:, :4].values # Assuming first 4 columns of df are features
y_target = df.iloc[:, 4].values.reshape(-1, 1) # Last column of df is the target

X_train, X_test, y_train, y_test = train_test_split(X_features, y_target, test_size=0.2, random_state=42)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(X_train_scaled.shape)
```

```python
class SGDRegressor():
  def __init__(self , learning_rate = 0.0001 , epochs = 1000):
    self.w = None
    self.b = None
    self.lr = learning_rate
    self.epochs = epochs

  def fit(self , X_train , y_train):
    self.b = 0.0 # Initialize as float
    self.w = np.ones(X_train.shape[1])
    for i in range(self.epochs):
      #In the entire dataset#
      #In batch we had to calculate the entire batch of data and then calculate the weight and bias#
      for j in range(X_train.shape[0]):
        #Just one row#
        idx = np.random.randint(0 , X_train.shape[0])
        #Generate a random row#

        current_X = X_train[idx]
        current_y = y_train[idx][0] # Access the scalar value for the current sample

        y_hat = np.dot(self.w , current_X) + self.b
        #This is the prediction value#

        # Correct gradient calculation using y_hat and current_y
        db = -2 * (current_y - y_hat)
        dw = -2 * (current_y - y_hat) * current_X

        self.b = self.b - self.lr * db
        self.w = self.w - self.lr * dw
         #Here we find the weight and bias for each#
        # print(self.w , self.b) # Commented out for cleaner output

  def predict(self , X_test):
    return np.dot(X_test , self.w) + self.b
     
```

```python
model = SGDRegressor()
model.fit(X_train_scaled , y_train) # Use scaled data
ypred = model.predict(X_test_scaled) # Use scaled data
# ypred # Remove this line to prevent printing ypred during execution
     

from sklearn.metrics import r2_score
r2_score_value = r2_score(y_test , ypred)
print(f"R2 Score: {r2_score_value}")
     
```
